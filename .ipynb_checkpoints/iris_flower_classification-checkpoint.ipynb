## Classifications of the iris flower types using support vector machine(SVM)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
columns = ['Sepal length', 'Sepal Width', 'Petal length', 'Petal width', 'Class_labels']
#load the data
df = pd.read_csv(r'C:\Users\User\Desktop\Data Science\Machine Learning\Data\iris.data', names = columns)
df.head()
```

```python
df.describe()
```

```python
sns.pairplot(df, hue='Class_labels')
```

```python
data = df.values
X = data[:, 0:4]
Y = data[:, 4]
```

```python
Y_Data = np.array([np.average(X[:, i][Y==j].astype('float32')) for i in range (X.shape[1])
                   for j in (np.unique(Y))])
Y_Data_reshaped = Y_Data.reshape(4,3)
Y_Data_reshaped = np.swapaxes(Y_Data_reshaped, 0, 1)
X_axis = np.arange(len(columns)-1)
width = 0.25
```

```python
plt.bar(X_axis, Y_Data_reshaped[0], width, label = 'Setosa')
plt.bar(X_axis+width, Y_Data_reshaped[1], width, label = 'Versicolour')
plt.bar(X_axis+width*2, Y_Data_reshaped[2], width, label = 'Virginica')
plt.xticks(X_axis, columns[:4])
plt.xlabel('Features')
plt.ylabel('Values in cm')
plt.legend(bbox_to_anchor=(1.3,1))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
from sklearn.svm import SVC
svn = SVC()
svn.fit(X_train, y_train)
```

```python
predictions =svn.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

## Testing the model

```python
X_new = [[3, 2, 1, 0.2],[4.9, 2.2, 3.5, 1.1],[5.3, 2.5, 4.6, 1.9]]
pred = svn.predict(X_new)
print(pred)
```

## Saving the model

```python
import pickle
with open('SVM.pickle', 'wb') as f:
    pickle.dump(svn, f)
```

## Loading the model

```python
with open('SVM.pickle', 'rb') as f:
    model = pickle.load(f)
model.predict(X_new)
```
